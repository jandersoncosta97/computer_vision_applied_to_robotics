# **Mapa de profundidade:**
 Avalie as métricas usando dois modelos de profundidade para 6 imagens, sendo metade indoor e metade outdoor.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.metrics import mean_squared_error, structural_similarity
import os

```

```python
root_path = '/content/drive/MyDrive/Visão Computacional Robótica/atividade_pratica/datasources/pictures/mapa_profundidade'
```

```python
for i in range(1, 7):
    print(os.path.join(root_path, f"pred_a_image_{i}.png"))
    if not os.path.exists(os.path.join(root_path, f"pred_a_image_{i}.png")):
        print("Doesn't exists")



```

```python

gt_depths = [cv2.imread(f"{root_path}/gt_image_{i}.jpg", cv2.IMREAD_GRAYSCALE) for i in range(1, 7)]
pred_depths_a = [cv2.imread(f"{root_path}/pred_a_image_{i}.jpg", cv2.IMREAD_GRAYSCALE) for i in range(1, 7)]
pred_depths_b = [cv2.imread(f"{root_path}/pred_b_image_{i}.jpg", cv2.IMREAD_GRAYSCALE) for i in range(1, 7)]
```

```python
image_labels = ["Outdoor"] * 3 + ["Indoor"] * 1 + ["Outdoor"] * 2
print(f"GT Depths: {[type(img) for img in gt_depths]}")
print(f"Pred A Depths: {[type(img) for img in pred_depths_a]}")
print(f"Pred B Depths: {[type(img) for img in pred_depths_b]}")
```

```python
def evaluate_depth_models(gt_depths, pred_depths_a, pred_depths_b, image_labels):
    results = {"Model_A": [], "Model_B": []}

    for i, (gt, pred_a, pred_b, label) in enumerate(zip(gt_depths, pred_depths_a, pred_depths_b, image_labels)):
        metrics_a = {
            "MAE": np.mean(np.abs(pred_a - gt)),
            "MSE": mean_squared_error(gt, pred_a),
            "SSIM": structural_similarity(gt, pred_a, data_range=gt.max() - gt.min())
        }
        metrics_b = {
            "MAE": np.mean(np.abs(pred_b - gt)),
            "MSE": mean_squared_error(gt, pred_b),
            "SSIM": structural_similarity(gt, pred_b, data_range=gt.max() - gt.min())
        }

        print(f"Image {i + 1} ({label}):")
        print(f"  Model A - MAE: {metrics_a['MAE']:.4f}, MSE: {metrics_a['MSE']:.4f}, SSIM: {metrics_a['SSIM']:.4f}")
        print(f"  Model B - MAE: {metrics_b['MAE']:.4f}, MSE: {metrics_b['MSE']:.4f}, SSIM: {metrics_b['SSIM']:.4f}")

        results["Model_A"].append(metrics_a)
        results["Model_B"].append(metrics_b)

    return results
```

```python
results = evaluate_depth_models(gt_depths, pred_depths_a, pred_depths_b, image_labels)
```

Não selecionei 3 imagens Indoors e 3 imagens Outdoors.
Os resultados indicam que não houveram nenhuma caso diferente de verdadeiro positivo. Isso ocorreu devido à organização do uso da imagem predita equivalente ao da imagem get_image_{i}.
